# Classification Analysis of Parkinson Speech Dataset 

In this project we are predicting Parkinsons Disease by using a publicly available dataset which is splited in train and test dataset.

In the training dataset, their is an extra column which is numerical value of Parkinson Disease level which is suggested by physians so that the dataset can be used for regression also. But in our project we are just trying to classify the parkinsons. So, we dont need that column in this project from the dataset. 

# Organizing out the Dataset

The dataset is not organized properly. For our requirements, we are just organizing it with some column names.

```python
import pandas as pd
columns = ["id"]
features = ["feature_"+str(i+1) for i in range(27)]
label = ["label"]

columns  = columns+features+label
df_train = pd.read_csv("train_data.txt", names = columns)
columns = columns[:27]+columns[28:]

df_test = pd.read_csv("test_data.txt", names = columns)
display(df_train.head())
display(df_test.head())
```

Dropping the column which contain numerical value to perform regression.

```python
df_train.drop('feature_27', axis=1, inplace=True)
```

```python
df_train.head()
```

Features Scaling and Setting the train, test data for training purpose and predicting.

```python
predictors_train = df_train.drop(['id'], axis = 1)
predictors_test = df_test.drop(['id'], axis = 1)


predictors_train = predictors_train.drop(['label'], axis = 1).values
predictors_test = predictors_test.drop(['label'], axis = 1).values

target_train = df_train['label']
target_test = df_test['label']

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler((-1, 1))
X_train = scaler.fit_transform(predictors_train)
X_test = scaler.fit_transform(predictors_test)

Y_train = target_train
Y_test = target_test
```

```python
X_train.shape, X_test.shape
```

# XGBoost Algorithm

```python
import xgboost as xgb
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
model = XGBClassifier()
model.fit(X_train, Y_train)

y_pred = model.predict(X_test)
print("XGB boost:")
print(metrics.accuracy_score(Y_test, y_pred))
print(metrics.classification_report(Y_test, y_pred))
print(metrics.confusion_matrix(Y_test, y_pred))
```

With XGBoost the accuracy is 68.45%. In case of predictiong there are zero accuracy while predicting non disease because in test data all of the instances from the persons who have parkinsons disease.

# Logistic Regression

```python
from sklearn import datasets
from sklearn import metrics
from sklearn.linear_model import LogisticRegression

# fit a logistic regression model to the data
model = LogisticRegression()
model.fit(X_train, Y_train)
y_pred = model.predict(X_test)
# summarize the fit of the model
print("Logistic Regression: ")
print(metrics.accuracy_score(Y_test, y_pred))
print(metrics.classification_report(Y_test, y_pred))
print(metrics.confusion_matrix(Y_test, y_pred))
```

With Logistic Regression the accuracy is 91.67%

# Gaussian Naive Bayes

```python
from sklearn import datasets
from sklearn import metrics
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X_train, Y_train)
# make predictions
y_pred = model.predict(X_test)
# summarize the fit of the model
print("Gaussian Naive Bayes:")
print(metrics.accuracy_score(Y_test, y_pred))
print(metrics.classification_report(Y_test, y_pred))
print(metrics.confusion_matrix(Y_test, y_pred))
```

With Gaussian Naive Bayes the accuracy is 85.12%

# k-Nearest Neighbor

```python
from sklearn import datasets
from sklearn import metrics
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier()
model.fit(X_train, Y_train)
# make predictions
y_pred = model.predict(X_test)
# summarize the fit of the model
print("k-Nearest Neighbor: ")
print(metrics.accuracy_score(Y_test, y_pred))
print(metrics.classification_report(Y_test, y_pred))
print(metrics.confusion_matrix(Y_test, y_pred))
```

With KNN the accuracy is 83.92%

# Support Vector Machine

```python
from sklearn import datasets
from sklearn import metrics
from sklearn.svm import SVC
model = SVC()
model.fit(X_train, Y_train)

# make predictions
y_pred = model.predict(X_test)
# summarize the fit of the model
print("Support Vector Machine: ")
print(metrics.accuracy_score(Y_test, y_pred))
print(metrics.classification_report(Y_test, y_pred))
print(metrics.confusion_matrix(Y_test, y_pred))
```

With SVM the accuracy is 62.5%

# Decision Tree

```python
from sklearn import datasets
from sklearn import metrics
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(X_train, Y_train)
# make predictions
y_pred = model.predict(X_test)
# summarize the fit of the model
print("Classification and Regression Trees")
print(metrics.accuracy_score(Y_test, y_pred))
print(metrics.classification_report(Y_test, y_pred))
print(metrics.confusion_matrix(Y_test, y_pred))
```
